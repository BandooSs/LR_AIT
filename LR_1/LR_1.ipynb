5. Решейпинг данных $1D\rightarrow2D$ с использованием `.pivot` (можно подать на вход результаты агрегатов, полученных ранее с использованием `.groupby` (сгруппировать по двум полям). 

```python
result = data.groupby("class")["cam_col"].count()# # Группируем данные по столбцу "class" и подсчитываем количество значений в столбце "cam_col" для каждой группы
result = result.reset_index()# Сбрасываем индекс
result = result.rename(columns={'cam_col': 'count_cam_col'})# Переименовываем столбец с количеством значений
pivot_df = result.pivot(index='class', columns='count_cam_col', values='count_cam_col')
pivot_df
```

7. Вычислить квантили распределения какого-либо вещественного признака (с использованием numpy.quantile или numpy.percentile).

```python
g_column = data["g"]# Извлекаем столбец "g"
q1 = np.percentile(g_column, 25)# Вычисляем первый квартиль (25-й перцентиль)
median = np.percentile(g_column, 50)# Вычисляем медиану (50-й перцентиль)
q3 = np.percentile(g_column, 75)# Вычисляем третий квартиль (75-й перцентиль)

print("Первый квантиль:", q1)
print("Медиана:", median)
print("Третий квантиль:", q3)
```
